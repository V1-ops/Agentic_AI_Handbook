Persistance is the abiliy to save or restore the state of a workflow over time 



```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from langgraph.checkpoint.memory import InMemorySaver
```

```python
load_dotenv()

llm = ChatOpenAI(model="gpt-4o-mini")
```

```python
class FlavourState(TypedDict):

    Dish: str
    Flavour: str
    explanation: str
```

```python
def generate_flavour(state: FlavourState):

    prompt = f'generate a flavour for the dish {state["Dish"]}'
    response = llm.invoke(prompt).content

    return {'Flavour': response}
```

```python
def generate_explanation(state: FlavourState):

    prompt = f'write an explanation for the flavour - {state["Flavour"]}'
    response = llm.invoke(prompt).content

    return {'explanation': response}
```

```python
graph = StateGraph(FlavourState)

graph.add_node('generate_flavour', generate_flavour)
graph.add_node('generate_explanation', generate_explanation)

graph.add_edge(START, 'generate_flavour')
graph.add_edge('generate_flavour', 'generate_explanation')
graph.add_edge('generate_explanation', END)

checkpointer = InMemorySaver()

workflow = graph.compile(checkpointer=checkpointer) # this tells langgraph that you want to store all the intermediate and final state results in the memory 
```

```python
config1 = {"configurable": {"thread_id": "1"}}
workflow.invoke({'Dish':'pizza'}, config=config1)
```

```python
workflow.get_state(config1)
```

```python
config2 = {"configurable": {"thread_id": "2"}}
workflow.invoke({'Dish':'brick'}, config=config2)
```
