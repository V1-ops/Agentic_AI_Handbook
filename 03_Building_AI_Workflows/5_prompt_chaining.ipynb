```python
from langgraph.graph import StateGraph, START,END
from typing import TypedDict
from langchain_huggingface import ChatHuggingFace,HuggingFaceEndpoint
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python

endpoint = HuggingFaceEndpoint(
    repo_id="Qwen/Qwen2.5-7B-Instruct",
    temperature=0.3,
    max_new_tokens=300,
)
llm = ChatHuggingFace(llm=endpoint)
```

```python
class blog_state(TypedDict):
    title: str
    outline : str
    content : str
```

```python
def get_outline(state:blog_state)->blog_state:
    # get title
    title = state['title']

    #based on title generate outline 
    outline_prompt = f'Based on the title{title},generate an outine for a blog'

    #get outline from the llm
    outline=llm.invoke(outline_prompt).content

    #update outline in the state 
    state['outline']= outline

    return state
```

```python
def generate_blog(state:blog_state)-> blog_state:

    title = state['title']
    outline = state['outline']

    prompt = f'write a detailed blog on title{title} and outline{outline}'

    content= llm.invoke(prompt).content

    state['content'] = content


    return state
```

```python
graph = StateGraph(blog_state)

graph.add_node('get_outline',get_outline)

graph.add_node('generate_blog',generate_blog)


graph.add_edge(START,'get_outline')
graph.add_edge('get_outline','generate_blog')
graph.add_edge('generate_blog',END)

workflow=graph.compile()
```

```python
initial_state = {'title': 'MOON a new home'}
final_state=workflow.invoke(initial_state)
print(final_state['content'])
```
